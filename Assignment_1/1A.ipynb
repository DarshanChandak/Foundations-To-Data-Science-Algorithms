**1-A) Prior and Posterior Distributions:**
---
Initial Beta Distribution: <br>
Given Parameters:<br> alpha = 2 and beta = 2.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta 
%matplotlib inline

#Initial Parameters given to us 
a,b = 2,2

#Using start x-index as 0 and end x-index as 1 
#200 points spaced at regular intervals to generate the plot below
x = np.linspace(0, 1, 200) 

plt.figure(figsize=(16,16))
plt.grid()

plt.xlim(0, 1) 
plt.xticks(np.arange(0, 1.1, 0.1))

plt.title('Beta Distribution', fontsize='20', color='b')
plt.ylabel('Probability Density', fontsize='14',color='r') 
plt.xlabel('Random Variable s', fontsize='14',color='r')

plt.plot(x, beta.pdf(x, a, b), 'green')
plt.show()
```

**1-A) Posterior distribution 1:**
---
It is said that in the initial survey, 40 out of 50 people liked the new update. Let p be the probability that a person who is randomly picked in the sample likes the 
new update. Then (1 - p) is the probability that he does not like the update. Of the given data, we can be sure that if we conduct this experiment 
50 times, we will get a positive outcome 40 times i.e. 40 of 50 people will like the new update.

**Likelihood function** indicates how likely a particular population is to produce an observed sample.
Thus, we calculate the likelihood of the particular population liking the new update.
The likelihood of random variable S becomes 40/50 i.e. 0.8.

Thus we add this new likelihood to our prior distribution (The last beta distribution). So, our total number of trial 
becomes (2 + 50) = 52, of which we can expect a positive outcome (2 + 40) = 42 times.

Hence, the parameters for the beta distribution change as follows: a = 2 + 40, b = 2 + 10.



```python
a,b = 42,12

#Using start x-index as 0 and end x-index as 1 
#200 points spaced at regular intervals to generate the plot below
x = np.linspace(0, 1, 200) 

plt.figure(figsize=(16,16))
plt.grid()

plt.xlim(0, 1)
plt.xticks(np.arange(0, 1.1, 0.1))

plt.title('Posterior distribution 1', fontsize='20',color='b' ) 
plt.ylabel('Probability Density', fontsize='14',color='r')
plt.xlabel('Random Variable s', fontsize='14',color='r')

plt.plot(x, beta.pdf(x, a, b), 'green')

#The below graph shows the Posterior Probability Distribution of a random variable S that is the probability of a customer liking a new update. 
#The graph is plottted after the 1st survey.
plt.show()
```

```python
df_test= df.sample(frac=0.2,random_state=49) 
df_train = df.drop(df_test.index)
y_test = df_test[['Appliances']]
y_test
X_test = df_test.drop(['Appliances'],axis=1)
X_test
y_train = df_train[['Appliances']]
y_train
X_train = df_train.drop(['Appliances'],axis=1)
X_train
```

```python
def correlation_matrix(dataset,relationship):
    correlation_matrix = dataset.corr()
    print(correlation_matrix)
    for i in correlation_matrix:
        relationship[i] = correlation_matrix[i][-1]
    
```

```python
relationship_sorted = sorted(relationship.items(), key = lambda x:abs(x[1]),reverse=True)
relationship_sorted

```

```python
variable_list = [x[0] for x in relationship_sorted]
variable_list


train_error_list = []
test_error_list= []
```

```python
def dataframe_to_matrix(dataframe):
    x = dataframe.to_numpy()
    return x
```

```python
def error_calculation(X_training,Y_training,theta):
    predicted = X_training.dot(theta)
    predicted_difference = predicted - Y_training
    predicted_difference_T = predicted_difference.T
    error = np.dot(predicted_difference_T,predicted_difference)
    error_rmse = np.sqrt(error/len(Y_training))
    return error_rmse
```

```python
def create_model(relationship_sorted,number_of_variables):
    column_required = variable_list[:number_of_variables]
    column_required.append('Appliances')
    new_dataframe = df_train[column_required]
    new_dataframe.insert(0,'intercept',1.0)
    new_dataframe2 = df_test[column_required]
    new_dataframe2.insert(0,'intercept',1.0)
    X_training_model = dataframe_to_matrix(new_dataframe.iloc[:,:-1])
    X_testing_model = dataframe_to_matrix(new_dataframe2.iloc[:,:-1])
    Y_testing_model = dataframe_to_matrix(new_dataframe2.iloc[:,-1])
    Y_training_model = dataframe_to_matrix(new_dataframe.iloc[:,-1])
    theta = np.linalg.inv(X_training_model.T@X_training_model)@X_training_model.T@Y_training_model
    print(theta)
    print("Error in training = {}".format(error_calculation(X_training_model,Y_training_model,theta)))
    train_error_val=error_calculation(X_training_model,Y_training_model,theta)
    print("Error in testing = {}".format(error_calculation(X_testing_model,Y_testing_model,theta)))
    test_error_val= error_calculation(X_testing_model,Y_testing_model,theta)
    train_error_list.append(train_error_val)
    test_error_list.append(test_error_val)
    print("\n\n\n\n\n")
    
```

```python
for i in range(26):
    create_model(relationship_sorted,i+1)
```

```python
x = list(range(2,28))
print(x)

print(train_error_list)
plt.figure(figsize=[12,12])
min_train_n = np.argmin(train_error_list)
min_test_n = np.argmin(test_error_list)
plt.plot(x,train_error_list, label="Training error values")
plt.plot(x,test_error_list, label="Testing error values")
plt.ylabel("Error Values")
plt.xlabel("Number of Variables")
plt.scatter(x[min_train_n], train_error_list[min_train_n], color="green", s=70)
plt.title("Plotting training and testing error of different models")
plt.legend()
plt.show()
```

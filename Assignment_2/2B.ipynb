```python
#importing python libraries
import numpy as np
import pandas as pd
import math as mp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dataframe_to_matrix(dataframe):
    x = dataframe.to_numpy()
    return x
```

```python
df_test= df.sample(frac=0.2,random_state=49) 
df_train = df.drop(df_test.index)
y_test = df_test[['Appliances']]
y_test
X_test = df_test.drop(['Appliances'],axis=1)
X_test
y_train = df_train[['Appliances']]
y_train
X_train = df_train.drop(['Appliances'],axis=1)
X_train
```

```python
variable_list = df.columns
variable_list = variable_list.drop('Appliances')
variable_list
```

```python
def error_calculation(X_training,Y_training,theta):
    predicted = X_training.dot(theta)
    predicted_difference = predicted - Y_training
    predicted_difference_T = predicted_difference.T
    error = np.dot(predicted_difference_T,predicted_difference)
    error_rmse = np.sqrt(error/len(Y_training))
    return error_rmse
```

```python
df_train.insert(0,'intercept',1.0)
df_test.insert(0,'intercept',1.0)
df_train = df_train.drop('rv2',axis=1)
df_train
```

```python
required_columns = [0]
error_min = mp.inf
print("Testing Error:")
for i in range(1,26):
    j_req = None
    for j in range(1,26):
        if(j not in required_columns):
            required_columns.append(j)
            X_training_model = dataframe_to_matrix(df_train.iloc[:,required_columns])
            Y_training_model = dataframe_to_matrix(df_train.iloc[:,-1]) 
            X_testing_model = dataframe_to_matrix(df_test.iloc[:,required_columns])
            Y_testing_model = dataframe_to_matrix(df_test.iloc[:,-1]) 
            theta = np.linalg.inv(X_training_model.T@X_training_model)@X_training_model.T@Y_training_model
            error_temp = error_calculation(X_testing_model,Y_testing_model,theta)
            error_temp_train = error_calculation(X_training_model,Y_training_model,theta)
            if(error_temp<error_min):
                error_min=error_temp
                j_req = j
            required_columns.pop()
    if(j_req==None):
        print("\n")
        print("Testing error:")
        print(error_min)
        print("Training error:")
        print(error_temp_train)
        break
    required_columns.append(j_req)
    print(error_min)
print(len(required_columns))
new_dataset_train = df_train.iloc[:,required_columns]
new_dataset_test = df_test.iloc[:,required_columns]
X_training_final = dataframe_to_matrix(new_dataset_train)
X_testing_final = dataframe_to_matrix(new_dataset_test)
Y_training_model = dataframe_to_matrix(df_train.iloc[:,-1]) 
Y_testing_model = dataframe_to_matrix(df_test.iloc[:,-1])
print(len(Y_testing_model))
```

```python
required_columns_backward = list(range(0,26)) # 0 to 25
error_min = mp.inf
print("Testing Error:")
for i in range(1,26):
    j_req = None
    j_value = 1010101
    
    for j in range(len(required_columns_backward)-1,-1,-1):
        if(j in required_columns_backward):
            j_value = required_columns_backward[j]
            required_columns_backward.pop(j)
            X_training_model = dataframe_to_matrix(df_train.iloc[:,required_columns_backward])
            Y_training_model = dataframe_to_matrix(df_train.iloc[:,-1]) 
            X_testing_model = dataframe_to_matrix(df_test.iloc[:,required_columns_backward])
            Y_testing_model = dataframe_to_matrix(df_test.iloc[:,-1]) 
            theta = np.linalg.inv(X_training_model.T@X_training_model)@X_training_model.T@Y_training_model
            error_temp = error_calculation(X_testing_model,Y_testing_model,theta)
            error_temp_train = error_calculation(X_training_model,Y_training_model,theta)
            if(error_temp<error_min):
                error_min=error_temp
                j_req = j_value
            required_columns_backward.append(j_value)
    if(j_req==None):
        print("\n")
        print("Testing error:")
        print(error_min)
        print("Training error:")
        print(error_temp_train)
        break
    for k in range(len(required_columns_backward)):
        if(required_columns_backward[k]==j_req):
            required_columns_backward.pop(k)
            break
    print(error_min)
print(len(required_columns_backward))
new_dataset_train = df_train.iloc[:,required_columns]
new_dataset_test = df_test.iloc[:,required_columns]
X_training_final = dataframe_to_matrix(new_dataset_train)
X_testing_final = dataframe_to_matrix(new_dataset_test)
Y_training_model = dataframe_to_matrix(df_train.iloc[:,-1]) 
Y_testing_model = dataframe_to_matrix(df_test.iloc[:,-1])
print(len(Y_testing_model))
```
